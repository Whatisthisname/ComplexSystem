```python
import bifurcation
import numpy as np

inv_temps = np.linspace(0.1, 2.0, 3 * 3 * 10)
runs = 10
data = np.zeros((len(inv_temps), runs))
for i_temp, inv_temp in enumerate(inv_temps):
    for i_run in range(runs):
        data[i_temp, i_run] = bifurcation.get_final_magnetization(
            inv_temp, seed=np.random.randint(0, 100000)
        )
```

```python
import matplotlib.pyplot as plt

for run in range(runs):
    plt.scatter(inv_temps, data[:, run], color="black", s=1)

plt.ylim(-1, 1)
plt.xlabel("Inverse temperature")
plt.ylabel("Steady state magnetization")
plt.show()
```
